```python
# Examine the distribution of unique categories in categorical columns
categorical_cols=df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns

for col in categorical_cols:
    print(f'Value counts for {col} :')
    display(df[col].value_counts())
```

## Univariate Analysis

```python
for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor)
```

## Bivariate Analysis

```python
for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns):
    plt.figure(i)
    sns.countplot(data=df, x=predictor,hue='Nationality')
```

```python
# Histoplot of value counts for different occupation

for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation count')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
```

## Numerical Analysis

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univariate anaalysis and visuaalization
plt.figure(figsize=(15,10))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)

plt.tight_layout(pad=1)
plt.show()
```

## Heatmaps

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

correlaton_matrix=df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlaton_matrix, annot=True, cmap=sns.cubehelix_palette(as_cmap=True))
plt.title('Correlaation Matrix')
plt.show()
```

```python
pairs_plot = [
    ('Bank Deposits','Saving Accounts'),
    ('Checking Accounts','Saving Accounts'),
    ('Checking Accounts','Foreign Currency Account'),
    ('Bank Loans', 'Credit Card Balance'),
    ('Business Lending', 'Bank Loans')
]
for x_col,y_col in pairs_plot:
    plt.figure(figsize=(8,6))
    sns.regplot(
        data=df,
        x=x_col,
        y=y_col,
        scatter_kws={'alpha':0.5},
        line_kws={'color':'red'}        
    )
    plt.title(f'Relationship between {x_col} and {y_col}',fontsize=14)
    plt.xlabel(x_col, fontsize=12)
    plt.ylabel(y_col, fontsize=12)
    plt.tight_layout()
    plt.show()
```

## Insights of EDA :

1. Strong Positive Correlation between Bank Deposits and Checking Accounts
   - Bank Deposits and Checking Accounts show a high correlation of around 0.84, This indicates customers with higher deposits also tend to maintain higher balances in checking accounts.

2. Saving Accounts are Highly Related to Bank Deposits
   - Saving Accounts and Bank Deposits have a strong positive correlation of approximately 0.75, This suggests customers actively saving money are also contributing significantly to bank deposits.

3. Moderate Relationship between Business Lending and Bank Loans
   - Business Lending and Bank Loans show a moderate correlation near 0.42, This implies customers taking business lending services are also likely to use loan products.

4. Most Features Show Positive Correlation
   - Almost all variables have positive correlations with each other, This indicates overall financial activities of customers are interconnected across banking services.
